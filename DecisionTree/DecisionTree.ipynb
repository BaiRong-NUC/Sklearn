- 决策树:监督学习算法,用于分类和回归任务,回归任务比较少见

- 决策树是一种树形结构
    树中的每个内部节点代表一个特征上的判断,每个分支代表一个判断结果的输出,每个叶子节点代表一种分类结果

- 决策树建立过程
    1. 特征选择: 选取有较强分类能力的特征作为决策树的节点
    2. 决策树生成: 根据选择的特征生成决策树
    3. 决策树采用剪枝技术来缓解过拟合

- 信息熵
    - 熵: 表示随机变量的不确定度的度量
        - 熵越大,随机变量的不确定度越大,信息越多
        - 熵越小,随机变量的不确定度越小
    - 信息熵: 公式= -Σ(分类占比*log2(分类占比)) 分类占比=某个分类的样本数/总样本数

- 信息增益
    - 特征a对训练数据集D的信息增益g(D,a),定义为集合D的熵H(D)与特征a给定条件下D的熵H(D|a)之差
    - 公式= g(D,a) = H(D) - H(D|a)
    - 条件熵H(D|a) = Σ(特征a的每个取值的占比*该取值下的熵)
    - 该取值下的熵= -Σ(该取值下的分类占比*log2(该取值下的分类占比)) 该取值下的分类占比=该取值下某个分类的样本数/该取值下的总样本数; 只看该取值下的样本,不是这个取值下的样本值忽略
    - 信息增益越大,说明特征a对数据集D的分类能力越强,越适合用来做决策树的节点

- ID3决策树构建:
    1. 计算每个特征的信息增益
    2. 使用信息增益最大的特征将数据集划分成子集
    3. 使用信息增益最大的特征作为决策树的节点
    4. 使用剩余特征对子集重复(1,2,3)步骤

- 信息增益率:
    - 信息增益率是信息增益的一种改进,用于解决ID3算法倾向于选择取值较多的特征作为节点的问题
    - 公式= 信息增益率(D,a) = g(D,a) / X 其中X=特征熵
    - 1/X: 惩罚系数
    - 特征熵X = -Σ(特征的每个取值的占比*log2(特征的每个取值的占比)) (注意:不看的样本的y,而是看特征值)
    - 信息增益率越大,说明特征a对数据集D的分类能力越强,越适合用来做决策树的节点

- C4.5决策树构建:
    1. 计算每个特征的信息增益率
    2. 使用信息增益率最大的特征将数据集划分成子集
    3. 使用信息增益率最大的特征作为决策树的节点
    4. 使用剩余特征对子集重复(1,2,3)步骤

- CART决策树:
    - CART是一种决策树模型,它既可以用于分类,也可以用于回归
    - CART回归树使用平方误差最小化策略
    - CART分类树使用基尼指数最小化策略

- 基尼系数:
    - 基尼值: 衡量数据集纯度的指标,基尼值越小,数据集越纯 
    - Gini(D) = 1 - Σ(分类占比^2)
    - 分类占比=某个分类的样本数/总样本数
    - 基尼系数 Gini(D,a) = Σ(特征a的每个取值的占比✖该取值下的基尼值)
        - 注意: 做二分划分,每个特征的取值只有两种,所以基尼系数的计算只需要考虑两个取值的情况 A 和非 A 
        - 如果特征连续,则需要将特征值进行升序排序,按照每两个相邻的特征值的平均值进行二分划分,计算每个划分的基尼系数,选择基尼系数最小的划分作为该特征的基尼系数
    - 基尼系数越小,说明特征a对数据集D的分类能力越强,越适合用来做决策树的节点
        - 当基尼系数相同时,选择特征值较少的特征作为节点,因为特征值较少的特征更容易解释和理解

- CART决策树构建:
    1. 计算每个特征的基尼系数
    2. 使用基尼系数最小的特征将数据集划分成子集
    3. 使用基尼系数最小的特征作为决策树的节点
    4. 使用剩余特征对子集重复(1,2,3)步骤

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier(
    criterion='gini',  # 基尼系数,也可以选择'entropy'信息增益
    max_depth=3,  # 树的最大深度
    min_samples_leaf=5,  # 叶子节点最少样本数
    min_samples_split=10,  # 内部节点再划分所需最小样本数
)
```

- CART回归树:
    - CART回归树使用平方误差最小化策略
    - 平方误差: 衡量回归模型预测值与真实值之间差异的指标,平方误差越小,模型的预测能力越强
    - 均方误差 MSE = 1/n * Σ(真实值 - 预测值)^2
    - 预测值 = 该节点下所有样本真实值的平均值
    - CART回归树构建流程:
        1. 选择一个特征,将该特征的值进行排序,取相邻点计算均值作为待划分点
        2. 根据所有划分点,将数据集分成两部分:R1、R2
        3. R1 和 R2 两部分的平方损失相加作为该切分点平方损失
        4. 取最小的平方损失的划分点,作为当前特征的划分点
        5. 以此计算其他特征的最优划分点,以及该划分点对应的损失值
        6. 在所有特征的划分点中,选择出最小平方损失的划分点,作为当前树的分裂点

- CART回归树容易出现过拟合的问题

- 决策树正则化:(剪枝)
    - 预剪枝: 在决策树生成过程中,提前停止树的生长,防止过拟合
        - 例如: 在生成子树之前,先评估子树的性能,如果子树的性能没有提升,则停止生长
    - 后剪枝: 在决策树生成完成后,对树进行修剪,去掉一些不必要的分支,提高模型的泛化能力
        - 例如: 使用交叉验证来评估每个子树的性能,选择性能最好的子树作为最终模型
    - 预剪枝容易出现欠拟合问题,后剪枝性能开销大
