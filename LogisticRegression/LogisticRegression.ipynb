- 逻辑回归: 适用于有监督二分类问题的线性模型，输出为概率值，常用于分类任务。
    - 数据->线性回归->预测值->激活函数->概率值->分类结果

- 激活函数: 将线性回归的输出转换为概率值，常用的激活函数包括Sigmoid函数和Softmax函数。
    - Sigmoid函数: 适用于二分类问题，将输入映射到0和1之间的概率值,结合阈值进行分类。
    - Softmax函数: 适用于多分类问题，将输入映射到多个类别的概率分布。

- 最大似然估计: 设发生事件的概率为p,不发生事件的概率为1-p,样本中有m个事件发生, n-m个事件不发生,则似然函数为L(p) = p^m * (1-p)^(n-m)。最大似然估计通过最大化似然函数来估计模型参数，使得观测数据在模型下的概率最大。

- 逻辑回归的损失函数: 
    - pi是逻辑回归的输出结果,每个样本有A,B两个类别,真实类别对应的位置概率越大越好
    - loss = - (y * log(pi) + (1-y) * log(1-pi))  # 二分类交叉熵损失函数
    - y是样本的真实标签,pi是模型预测的概率值,当y=1时,损失函数为-log(pi),当y=0时,损失函数为-log(1-pi)。通过最小化损失函数来优化模型参数。
    - 求解损失函数的最小值可以使用梯度下降算法,通过计算损失函数关于模型参数的梯度,不断更新参数直到收敛。和线性回归的梯度下降类似
